# Data Visualization using Train dataset.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.drop('Cabin',axis=1, inplace=True)
```

```python
df.drop('PassengerId',axis=1, inplace=True)
```

```python
df.drop('Name', axis=1, inplace=True)
```

```python
df.head()
```

```python
#counts the number of unique items in the column and plots it
sns.countplot(df['Pclass'])
```

```python
sns.countplot(df['Pclass'], hue=df['Sex'])
```

```python
sns.scatterplot(df['Age'])
```

```python
df['Embarked'].unique()
```

```python
sns.countplot(df['Embarked'], hue=df['Survived'])
```

Categorical scatterplots:

stripplot() (with kind="strip"; the default)

swarmplot() (with kind="swarm")

Categorical distribution plots:

boxplot() (with kind="box")

violinplot() (with kind="violin")

boxenplot() (with kind="boxen")

Categorical estimate plots:

pointplot() (with kind="point")

barplot() (with kind="bar")

countplot() (with kind="count")

```python
#catplot() is a single function used to plot different kinds of graph mentioned above by changing the value of kind var.
sns.catplot(y=df['Age'], x=df['Fare'], data=df)
```
